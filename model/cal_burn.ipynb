```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import  seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
df1=pd.read_csv(r"C:\Users\ravin\Downloads\calories_burnnnn\calories.csv")
df2=pd.read_csv(r"C:\Users\ravin\Downloads\calories_burnnnn\exercise.csv")
df3=pd.concat([df2,df1['Calories']],axis=1)
#df3.replace({"Gender":{'male':0,'female':1}}, inplace=True)
```

```python
dummies=pd.get_dummies(df3.Gender)
dummies.head(3)
```

```python
dummies = dummies.astype(int)
dummies.head(3)
```

```python
df3=pd.concat([df3,dummies],axis=1)
df3.tail(3)
```

```python
df3=df3.drop(['Gender'],axis=1)
df3
```

```python
X = df3.drop(['User_ID','Calories'], axis=1)
Y = df3['Calories']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=2)
model = XGBRegressor()
model.fit(X_train, Y_train)
test_data_pred = model.predict(X_test)
mae = metrics.mean_absolute_error(Y_test, test_data_pred)
print(mae)
print(model.score(X_test, Y_test))
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
crs_validation=ShuffleSplit(n_splits=5, test_size=0.25, random_state=20)
print(cross_val_score(XGBRegressor(),X,Y, cv=crs_validation))
```

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
import pandas as pd

def find_best_model_using_gridsearchcv(X, Y):
    algos = {
        'xgboost': {
            'model': XGBRegressor(),
            'params': {
                'n_estimators': [100, 200, 300],  # Add more parameters as needed
                'learning_rate': [0.01, 0.1, 0.2],
                'max_depth': [3, 4, 5],
                # Add other XGBRegressor parameters you want to tune
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }

    scores = []

    crs_val = ShuffleSplit(n_splits=5, test_size=0.2, random_state=114433)

    for algo_name, algo in algos.items():
        gscv = GridSearchCV(algo['model'], algo['params'], cv=crs_val, return_train_score=False)
        gscv.fit(X, Y)
        scores.append({
            'algorithm': algo_name,
            'best_params': gscv.best_params_,
            'best_score': gscv.best_score_
        })

    return pd.DataFrame(scores, columns=['algorithm', 'best_score', 'best_params'])

# Assuming x and y are defined elsewhere
find_best_model_using_gridsearchcv(X, Y)
```

```python
def predict_cal(Gender,Age,Height,Weight,Duration,Heart_Rate,Body_Temp):
    loc_index = np.where(X.columns==Gender)[0][0]
    p=np.zeros(len(X.columns))
    p[0]=Age
    p[1]=Height
    p[2]=Weight
    p[3]=Duration
    p[4]=Heart_Rate
    p[5]=Body_Temp
    if loc_index>=0:
        p[loc_index]=1
    return model.predict([p])[0]
```
